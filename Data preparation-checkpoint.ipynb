```python
from PIL import Image
import os

def crop_image(image_path, crop_size):
    image = Image.open(image_path)
    width, height = image.size
    crops = []

    # Check if the image dimensions are divisible by crop size
#     if width % crop_size != 0 or height % crop_size != 0:
#         print("Image dimensions are not divisible by the crop size.")
#         return crops

    # Crop the image into multiple smaller images
    for i in range(0, width, crop_size):
        for j in range(0, height, crop_size):
            box = (i, j, i + crop_size, j + crop_size)
            cropped_image = image.crop(box)
            crops.append(cropped_image)

    return crops

def save_cropped_images(crops, output_prefix):
    for i, crop in enumerate(crops):
        output_path = f"./cropped/{output_prefix}_{i+1}.png"
        crop.save(output_path)
        print(f"Saved {output_path}")

def get_filenames_in_path(directory):
    filenames = []
    for filename in os.listdir(directory):
        if os.path.isfile(os.path.join(directory, filename)):
            filenames.append(filename)
    return filenames
```

```python
directory_path = ".\large"
filenames = get_filenames_in_path(directory_path)
```

```python
# Example usage
for filename in filenames:
    image_path = ".\large\\" + str(filename)
    crop_size = 1024
    output_prefix = filename[:-4]
    crops = crop_image(image_path, crop_size)
    save_cropped_images(crops, output_prefix)
```

```python
from PIL import Image
import os

def compress_images(directory, output_directory, target_size=(256, 256)):
    for filename in os.listdir(directory):
        if filename.endswith((".jpg", ".tif", ".png")):
            image_path = os.path.join(directory, filename)
            output_path = os.path.join(output_directory, filename)
            compress_image(image_path, output_path, target_size)

def compress_image(image_path, output_path, target_size):
    image = Image.open(image_path)
    image.thumbnail(target_size)
    image.save(output_path)
    print(f"Compressed {image_path} and saved as {output_path}")

# Example usage
input_directory = "./large"
output_directory = "./256_256"

compress_images(input_directory, output_directory)
```
